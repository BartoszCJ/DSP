```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram
```

```python
N = 1000 


uniform_noise = np.random.uniform(low=-1.0, high=1.0, size=N)


gaussian_noise = np.random.normal(loc=0.0, scale=1.0, size=N)
```

```python

mean_uniform = np.mean(uniform_noise)
mean_gaussian = np.mean(gaussian_noise)

var_uniform = np.var(uniform_noise)
var_gaussian = np.var(gaussian_noise)

print(f"Uniform Noise - Mean: {mean_uniform:.4f}, Variance: {var_uniform:.4f}")
print(f"Gaussian Noise - Mean: {mean_gaussian:.4f}, Variance: {var_gaussian:.4f}")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(uniform_noise)
plt.title("Uniform Noise")
plt.xlabel("Sample")
plt.ylabel("Amplitude")

plt.subplot(1, 2, 2)
plt.plot(gaussian_noise)
plt.title("Gaussian Noise")
plt.xlabel("Sample")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```

```python
fs = 1000 

f_uni, Pxx_uni = periodogram(uniform_noise, fs)

f_gauss, Pxx_gauss = periodogram(gaussian_noise, fs)

plt.figure(figsize=(12, 5))

plt.semilogy(f_uni, Pxx_uni, label='Uniform Noise')
plt.semilogy(f_gauss, Pxx_gauss, label='Gaussian Noise', linestyle='--')

plt.title("PSD Comparison of Uniform and Gaussian Noise")
plt.xlabel("Frequency [Hz]")
plt.ylabel("PSD [dB/Hz]")
plt.legend()
plt.grid(True)
plt.show()
```

```python
f_u, Pxx_u = periodogram(uniform_noise,  fs=fs, scaling='density')
f_g, Pxx_g = periodogram(gaussian_noise, fs=fs, scaling='density')
```
