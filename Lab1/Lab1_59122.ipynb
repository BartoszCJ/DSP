```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
X_mu = np.array([6, 8, 2, 4, 3, 4, 5, 0, 0, 0], dtype=complex)
```

```python
k  = np.arange(N).reshape(-1, 1)
mu = np.arange(N).reshape(1,  -1)
K  = k @ mu
WN = np.exp(2j * np.pi / N)
W  = WN ** K
x_k = (1 / N) * W @ X_mu
```

```python
print("x[k] =", np.real_if_close(x_k))
```

```python
plt.figure()

plt.stem(np.arange(N), np.real(x_k), basefmt=" ")
plt.xlabel("k")
plt.ylabel("Amplitude")
plt.tight_layout()


plt.show()
```
